```python
import pandas as pd
from pandas import read_excel

dfAtletas = pd.read_excel("/content/Athletes.xlsx", index_col= False) 
dfTreinadores = pd.read_excel("/content/Coaches.xlsx", index_col= False) 
dfSexo = pd.read_excel("/content/EntriesGender.xlsx", index_col= False) 
dfMedalhas = pd.read_excel("/content/Medals.xlsx", index_col= False) 
dfTimes= pd.read_excel("/content/Teams.xlsx", index_col= False) 

dfAtletas = dfAtletas.applymap(lambda s:s.upper() if type(s) == str else s)
dfTreinadores = dfTreinadores.applymap(lambda s:s.upper() if type(s) == str else s)
dfSexo = dfSexo.applymap(lambda s:s.upper() if type(s) == str else s)
dfMedalhas = dfMedalhas.applymap(lambda s:s.upper() if type(s) == str else s)
dfTimes = dfTimes.applymap(lambda s:s.upper() if type(s) == str else s)

# criando df de atletas por esporte  totalPorEsporte():
dfSexo["Total"] = dfSexo['Male'] + dfSexo['Female']
dfTotalPorEsporte = dfSexo
dfTotalPorEsporte = dfTotalPorEsporte.drop(columns=['Female'])
dfTotalPorEsporte = dfTotalPorEsporte.drop(columns=['Male'])

# criando df total de medalhas por pais 
dfMedalhas["Total"] = dfMedalhas['Gold'] + dfMedalhas['Silver'] + dfMedalhas['Bronze']
dfTotalMedalhasPorPais = dfMedalhas
dfTotalMedalhasPorPais = dfTotalMedalhasPorPais.drop(columns=['Gold'])
dfTotalMedalhasPorPais = dfTotalMedalhasPorPais.drop(columns=['Silver'])
dfTotalMedalhasPorPais = dfTotalMedalhasPorPais.drop(columns=['Bronze'])
dfTotalMedalhasPorPais = dfTotalMedalhasPorPais.drop(columns=['Rank'])

# criando df ranking por total de medalas 

dfRankingPorTotalMedalhas =dfTotalMedalhasPorPais.sort_values(by=['Total'], ignore_index=True, ascending=False)

# criando df pais com mais medalhas de Ouro():
dfMaiorOuro = dfMedalhas[dfMedalhas['Gold'] == dfMedalhas['Gold'].max()]
dfMaiorOuro = dfMaiorOuro.drop(columns=['Silver'])
dfMaiorOuro = dfMaiorOuro.drop(columns=['Bronze'])
dfMaiorOuro = dfMaiorOuro.drop(columns=['Total'])
dfMaiorOuro = dfMaiorOuro.drop(columns=['Rank'])

# criando df pais com mais medalhas de prata 
dfMaiorPrata = dfMedalhas[dfMedalhas['Silver'] == dfMedalhas['Silver'].max()]
dfMaiorPrata = dfMaiorPrata.drop(columns=['Gold'])
dfMaiorPrata = dfMaiorPrata.drop(columns=['Bronze'])
dfMaiorPrata = dfMaiorPrata.drop(columns=['Total'])
dfMaiorPrata = dfMaiorPrata.drop(columns=['Rank'])

# criando df pais com mais  medalhas de bronze 
dfMaiorBronze = dfMedalhas[dfMedalhas['Bronze'] == dfMedalhas['Bronze'].max()]
dfMaiorBronze = dfMaiorBronze.drop(columns=['Silver'])
dfMaiorBronze = dfMaiorBronze.drop(columns=['Gold'])
dfMaiorBronze = dfMaiorBronze.drop(columns=['Total'])
dfMaiorBronze = dfMaiorBronze.drop(columns=['Rank'])

# criando df pais com menos medalhas de ouro 
dfMenorOuro = dfMedalhas[dfMedalhas['Gold'] == dfMedalhas['Gold'].min()]
dfMenorOuro = dfMenorOuro.drop(columns=['Silver'])
dfMenorOuro = dfMenorOuro.drop(columns=['Bronze'])
dfMenorOuro = dfMenorOuro.drop(columns=['Total'])
dfMenorOuro = dfMenorOuro.drop(columns=['Rank']) 


# criando df pais com menos medalhas de prata  
dfMenorPrata = dfMedalhas[dfMedalhas['Silver'] == dfMedalhas['Silver'].min()]
dfMenorPrata = dfMenorPrata.drop(columns=['Gold'])
dfMenorPrata = dfMenorPrata.drop(columns=['Bronze'])
dfMenorPrata = dfMenorPrata.drop(columns=['Total'])
dfMenorPrata = dfMenorPrata.drop(columns=['Rank'])

# criando df pais com menos medalhas de bronze 
dfMenorBronze = dfMedalhas[dfMedalhas['Bronze'] == dfMedalhas['Bronze'].min()]
dfMenorBronze= dfMenorBronze.drop(columns=['Silver'])
dfMenorBronze = dfMenorBronze.drop(columns=['Gold'])
dfMenorBronze = dfMenorBronze.drop(columns=['Total'])
dfMenorBronze = dfMenorBronze.drop(columns=['Rank'])

# criando df com esportes participantes 
dfEsportes =  dfSexo['Discipline']

# criando df esporte com mais homens 
dfEsporteComMaisHomens = dfSexo.query('Male > Female ')
dfEsporteComMaisHomens = dfEsporteComMaisHomens.drop(columns=['Total'])

# criando df esporde com mais mulheres   
dfEsporteComMaisMulheres = dfSexo.query('Male < Female ')
dfEsporteComMaisMulheres = dfEsporteComMaisMulheres.drop(columns=['Total'])

# criando df quandidade treinadores por pais 
dfTreinadoresPorPais = dfTreinadores['NOC'].value_counts()

#criando df quantidade treinadores por esporte 
dfTreinadoresPorEsporte = dfTreinadores['Discipline'].value_counts()

#criando df quantidade de times por esporte de cada pais 
dfTimesPorEsporte = dfTimes['Name'].value_counts()  

# funcoes 
def totalAtletas():
  print(len(dfAtletas.index))

def totalHomens():
  print(sum(dfSexo['Male']))

def totalMulheres(): 
  print(sum(dfSexo['Female']))

  
def paisMaisTreinadores():
  treinadores = dfTreinadoresPorPais
  listaTreinadores =treinadores.to_dict()
  for chave, valor in listaTreinadores.items():
    pais = chave
    treinadores = valor
    break 
  print(f"{pais}, com {treinadores} treinadores")

def informacoesBrasil():
  print('ATLETAS BRASILEIROS')
  print((dfAtletas[dfAtletas['NOC']  == 'BRAZIL']).to_string())

  print('TREINADORES BRASILEIROS')
  print((dfTreinadores[dfTreinadores['NOC']  == 'BRAZIL']).to_string())

  print('MEDALHAS BRASILEIROS')
  print((dfMedalhas[dfMedalhas['Team/NOC']  == 'BRAZIL']).to_string())

  print('TIMES BRASILEIROS BRASILEIROS')
  print((dfTimes[dfTimes['NOC']  == 'BRAZIL']).to_string())

def informacoesPaisEscolhidoPeloUsuario():
  pais = str(input('digite o pais que deseja saber todas informações')).upper()
  teste = (dfMedalhas[dfMedalhas['Team/NOC']  == pais])
  while teste.empty: 
    print("pais invalido ou nao participou das olimpiadas digite outro: ")
    pais = str(input()).upper()
    teste = (dfMedalhas[dfMedalhas['Team/NOC']  == pais])

  print(f'ATLETAS d@ {pais}')
  print((dfAtletas[dfAtletas['NOC']  == pais]).to_string())

  print(f'TREINADORES d@ {pais}')
  print((dfTreinadores[dfTreinadores['NOC']  == pais]).to_string())

  print(f'MEDALHAS d@ {pais}')
  print((dfMedalhas[dfMedalhas['Team/NOC']  == pais]).to_string())

  print(f'TIMES d@ {pais}')
  print((dfTimes[dfTimes['NOC']  == pais]).to_string())

# mensagens

mensagemTotalAtletasParticipante = " 1- Total de atletas participantes\n"
mensagemTotalHomens = "2- Total de participantes homens\n"
mensagemTotalMulheres = "3- Total de participantes mulheres\n"
mensagemTotalPorEsporte =  "4- Total de participantes por esporte\n"
mensagemMedalhasPorPais = "5- Total de medalhas por país\n"
mensagemRankingPorMedalhas = "6- Ranking por medalhas totais\n"
mensagemPaisComMaisOuro =  "7- País com mais medalhas de ouro\n"
mensagemPaisComMaisPrata = "8- País com mais medalhas de prata\n"
mensagemPaisComMaisBronze = "9- País com mais medalhas de bronze\n"
mensagemPaisComMenosOuro = "10- País com menos medalhas de ouro\n"
mensagemPaisComMenosPrata = "11- País com menos medalhas de prata\n"
mensagemPaisComMenoSBronze = "12- País com menos medalhas de bronze\n"
mensagemListaEsportesParticipantes = "13- Lista com esportes participante\n"
mensagemListaEsporesComMaisHomens = "14- Lista de esportes com mais homens que mulheres\n"
mensagemListaEsporesComMaisMulheres = "15- Lista de esportes com mais mulheres que homens\n"
mensagemTreinadoresPorPais ="16- Quantidade de treinadores por país\n"
mensagemPaisComMaisTreinadores = "17- País com a maior quantidade de treinadores\n"
mensagemTreinadoresPorEsporte = "18- Quantidade de treinadores por esporte\n"
mensagemTimesPorEsporte = "19- Quanto times por esporte cada país tem\n"
mensagemTodosDadosBrazil = "20- Todos os dados sobre o brazil\n"
mensagemTodosDadosPaisEscolhido  ="21- Todos os dados sobre um pais a sua escolha\n"

# inicio sistema 
print("Bem-Vind@ ao sistema de analise de dados das olimpiadas de Tokio 2021")
print("Qual dos dados abaixo voce quer obter?\n")
print(mensagemTotalAtletasParticipante, mensagemTotalHomens, mensagemTotalMulheres,
      mensagemTotalPorEsporte, mensagemMedalhasPorPais, mensagemRankingPorMedalhas,
      mensagemPaisComMaisOuro, mensagemPaisComMaisPrata, mensagemPaisComMaisBronze,
      mensagemPaisComMenosOuro, mensagemPaisComMenosPrata, mensagemPaisComMenoSBronze, 
      mensagemListaEsportesParticipantes, mensagemListaEsporesComMaisHomens, 
      mensagemListaEsporesComMaisMulheres, mensagemTreinadoresPorPais, 
      mensagemPaisComMaisTreinadores, mensagemTreinadoresPorEsporte, mensagemTimesPorEsporte, 
      mensagemTodosDadosBrazil, mensagemTodosDadosPaisEscolhido)
aux = 'SIM'
while True:
  if aux == 'SIM' or aux == '1':
    while True:
      try:
        opcao = int(input('escolha um numero de 1 a 21: '))
        break
      except ValueError:
         print("Somente numeros sao aceitos. Tente novamente.")

    while (opcao < 1 or opcao >23): 
      print("NUMERO INVALIDO \ndigite seu numero de 1 e 21 para obter os dados: ")
      opcao = int(input())
    if opcao == 1:
      print("1- Total de atletas participantes.\n")
      totalAtletas()
     
    elif opcao == 2:
      print(mensagemTotalHomens)
      totalHomens()

    elif opcao == 3:  
      print(mensagemTotalMulheres)
      totalMulheres()

    elif opcao == 4:
      print(mensagemTotalPorEsporte)
      print(dfTotalPorEsporte)

    elif opcao == 5:  
      print(mensagemMedalhasPorPais)
      print(dfTotalMedalhasPorPais) 

    elif opcao == 6: 
      print(mensagemRankingPorMedalhas)  
      print(dfRankingPorTotalMedalhas)

    elif opcao == 7:
      print(mensagemPaisComMaisOuro)
      print(dfMaiorOuro)

    elif opcao == 8:  
      print(mensagemPaisComMaisPrata)
      print(dfMaiorPrata)

    elif opcao == 9:
      print(mensagemPaisComMaisBronze)
      print(dfMaiorBronze)

    elif opcao == 10:   
      print(mensagemPaisComMenosOuro)
      print(dfMenorOuro)

    elif opcao == 11:
      print(mensagemPaisComMenosPrata)
      print(dfMenorPrata)

    elif opcao == 12:  
      print(mensagemPaisComMenoSBronze)
      print(dfMenorBronze)

    elif opcao == 13:
      print(mensagemListaEsportesParticipantes)
      print(dfEsportes)      

    elif opcao == 14:
      print(mensagemListaEsporesComMaisHomens)
      print(dfEsporteComMaisHomens)

    elif opcao == 15:  
      print(mensagemListaEsporesComMaisMulheres)
      print(dfEsporteComMaisMulheres)

    elif opcao == 16:
      print(mensagemTreinadoresPorPais)
      print(dfTreinadoresPorPais)

    elif opcao == 17: 
      print(mensagemPaisComMaisTreinadores)
      paisMaisTreinadores()

    elif opcao == 18:  
      print(mensagemTreinadoresPorEsporte)
      print(dfTreinadoresPorEsporte)

    elif opcao == 19:
      print(mensagemTimesPorEsporte)
      print(dfTimesPorEsporte)

    elif opcao == 20: 
      print(mensagemTodosDadosBrazil)
      informacoesBrasil()  

    elif opcao == 21:
      print(mensagemTodosDadosPaisEscolhido)
      informacoesPaisEscolhidoPeloUsuario()

    aux = str(input('\nDigite sim ou 1 se deseja ver outro dado e qualquer outra coisa para sair: ')).upper()
  else:
      print('\n\nObrigada por ter usado o sistema')
      break










```
